```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=pd.read_csv('junta_seca.csv')
b=pd.read_csv('solda.csv')
c=pd.read_csv('squat.csv')
```

```python
plt.figure(figsize=[20,10])
plt.subplot(311)
plt.plot(a.iloc[:50,:-1].T,color='b',alpha=0.08);
plt.grid(True)
plt.xticks(np.arange(0, len(a.iloc[0,:-1].T)+1, 50))
ax = plt.gca()
ax.set_xticklabels([]);
plt.subplot(312)
plt.plot(b.iloc[:50,:-1].T,color='r',alpha=0.08);
plt.grid(True)
plt.xticks(np.arange(0, len(b.iloc[0,:-1].T)+1, 50))
ax = plt.gca()
ax.set_xticklabels([]);
plt.subplot(313)
plt.plot(c.iloc[:50,:-1].T,color='c',alpha=0.08);
plt.grid(True)
plt.xticks(np.arange(0, len(c.iloc[0,:-1].T)+1, 50))
ax = plt.gca()
ax.set_xticklabels([]);
```

```python
plt.figure(figsize=[20,10])
plt.subplot(311)
plt.plot(a.iloc[:,:-1].mean().values,'b')
plt.grid(True)
plt.subplot(312)
plt.plot(b.iloc[:,:-1].mean().values,'r')
plt.grid(True)
plt.subplot(313)
plt.plot(c.iloc[:,:-1].mean().values,'c')
plt.grid(True)
```

```python
df=a.append(b).append(c)
```

```python
import sklearn as sk
```

```python
from sklearn.model_selection import train_test_split as split
```

```python
x_train,x_test,y_train,y_test=split(df.iloc[:,:-1],df.TARGET)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.ensemble import ExtraTreesClassifier as xtree
```

```python
model=xtree()
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, model.predict(x_test))
```
