```python
import plotly.express as plt
import pandas as pd
```

```python
df48=pd.DataFrame(data48)
```

```python
cols=[]
for i,x in enumerate(['hv']*8):
    cols.append(x[0]+str(i+1))
    cols.append(x[1]+str(i+1))
```

```python
df48.columns=cols
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
```

```python
canales=['canal_'+str(i+1) for i in range(8)]
```

```python
p = figure(x_axis_label='h', y_axis_label='v')
for i in range(int(len(cols)/2)):
    p.line(df48.iloc[:,2*i], df48.iloc[:,2*i+1], legend_label=canales[i], line_width=2)
show(p)
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=1)

for i in range(int(len(cols)/2)):
    fig.add_trace(
    go.Scatter(x=df48.iloc[:,2*i], y=df48.iloc[:,2*i+1], name=canales[i]))

fig.update_layout(height=600, width=1000)
fig.show()
```
